```python
import numpy as np
import matplotlib.pyplot as plt

# Lataa molemmat mallit ja niiden testitulokset
cnn_data = np.load('age_CNN_history_with_test.npz', allow_pickle=True)
mobilenet_data = np.load('age_MobileNet_history_with_test.npz', allow_pickle=True)
```

```python
def compare_models(npz_paths, labels, title="Training & Validation Accuracy"):
    plt.figure(figsize=(10, 6))
    for path, label in zip(npz_paths, labels):
        data = np.load(path, allow_pickle=True)
        history = data['history'].item()
        test_acc = data['test_acc'].item()
        plt.plot(history['accuracy'], label=f'{label} Train')
        plt.plot(history['val_accuracy'], label=f'{label} Val')
        print(f"{label} test accuracy: {test_acc:.2%}")
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


compare_models(
    ["age_CNN_history.npz", "age_MobileNet_history.npz"],
    ["CNN", "MobileNet"]
)
```
